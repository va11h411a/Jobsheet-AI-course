```python
#Tugas Praktikum - 5

import pandas as pd

# Load the dataset
df = pd.read_csv('/content/epl-goalScorer(20-21).csv')

# Drop the 'Unnamed: 0' column if it exists
df.drop(columns=['Unnamed: 0'], inplace=True)

# 5. Display the total number of goals scored by each player
goals_per_player = df[['player_name', 'goals']].groupby('player_name').sum()
print("Jumlah goal yang dicetak setiap pemain:")
print(goals_per_player)
```

```python
#Tugas Praktikum - 6
# 6. Display a DataFrame consisting of player name, goals, position, team name, and yellow cards
player_stats = df[['player_name', 'goals', 'position', 'team_title', 'yellow_cards']]
print("\nDataFrame pemain dengan nama, goal, posisi, klub, dan jumlah kartu kuning:")
print(player_stats)
```

```python
#Tugas Praktikum - 7
# 7. Display all players from the "Liverpool" club
liverpool_players = df[df['team_title'] == 'Liverpool']
print("\nPemain dari klub 'Liverpool':")
print(liverpool_players[['player_name', 'team_title']])
```

```python
#Tugas Praktikum - 8
# 8. Display all players from the "Arsenal" club
arsenal_players = df[df['team_title'] == 'Arsenal']
print("\nPemain dari klub 'Arsenal':")
print(arsenal_players[['player_name', 'team_title']])
```

```python
#Tugas Praktikum - 9
# 9. Display the top 10 players with the most yellow cards
top_yellow_cards = df[['player_name', 'yellow_cards']].sort_values(by='yellow_cards', ascending=False).head(10)
print("\n10 pemain dengan jumlah kartu kuning terbanyak:")
print(top_yellow_cards)
```

```python
#Tugas Praktikum - 10
# 10. Display the top 10 players with the most red cards
top_red_cards = df[['player_name', 'red_cards']].sort_values(by='red_cards', ascending=False).head(10)
print("\n10 pemain dengan jumlah kartu merah terbanyak:")
print(top_red_cards)
```
