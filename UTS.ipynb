## Manipulation

```python
df_airbnb['last_review'] = pd.to_datetime(df_airbnb['last_review'])
df_manipulation = df_airbnb.dropna(subset=['last_review']).copy()
print(f"Jumlah baris setelah menghapus null di 'last_review': {len(df_manipulation)}")
df_manipulation.head(10)
```

```python
df_manipulation['review_year'] = df_manipulation['last_review'].dt.year
df_manipulation['review_month'] = df_manipulation['last_review'].dt.month
df_manipulation.head(10)
```

```python
df_cleaned = df_manipulation[df_manipulation['availability_365'] != 0].copy()
df_cleaned.head(10)
```

```python
df_data = df_cleaned[['price', 'minimum_nights', 	'number_of_reviews', 'reviews_per_month',	'calculated_host_listings_count',	'availability_365']].describe()
df_data
```

```python
hosts_per_area = df_cleaned.groupby('neighbourhood_group')['host_id'].nunique().reset_index()
hosts_per_area.columns = ['Neighbourhood Group', 'Jumlah pengguna']

print(hosts_per_area)
```

## Visualization

```python
room_type_counts = df_cleaned['room_type'].value_counts()


plt.figure(figsize=(8, 8))
plt.pie(
    room_type_counts.values,
    labels=room_type_counts.index,
    autopct='%1.1f%%',
    colors=['#FFD65A', '#F93827', '#16C47F'],
    explode = (0, 0, 0.1),
    startangle=140,
    shadow=True
)

plt.title('jumlah data ')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


avg_price = df_cleaned.groupby('neighbourhood_group')['price'].mean()

plt.plot(avg_price.index, avg_price.values, marker='o', linestyle='-', color='blue')

plt.title('Rata-rata Harga per Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Rata-rata Harga')
plt.grid(True)

plt.show()
```

```python
custom_palette = {
    'Brooklyn': '#1f77b4',
    'Manhattan': '#ff7f0e',
    'Queens': '#2ca02c',
    'Bronx': '#d62728',
    'Staten Island': '#9467bd'
}

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(
    data=hosts_per_area,
    y='Neighbourhood Group',
    x='Jumlah pengguna',
    palette=custom_palette,
    orient='h'
)

# Menambahkan judul dan label
plt.title('Jumlah pengguna per Neighborhood Group', pad=20, fontsize=16)
plt.xlabel('Jumlah penggunna', labelpad=10)
plt.ylabel('Neighborhood Group', labelpad=10)
for index, value in enumerate(hosts_per_area['Jumlah pengguna']):
    plt.text(
        value + 5,
        index,
        f"{value:,}",
        va='center',
        fontsize=10
    )

plt.tight_layout()
plt.show()
```

```python
neighborhood_prices = df_cleaned.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
neighborhood_prices.head(20).plot(kind='bar')
plt.title('Rata-rata Harga per Neighborhood (Top 20)')
plt.ylabel('Harga Rata-rata ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
from calendar import month_abbr
df_cleaned['review_month'] = df_cleaned['last_review'].dt.month


monthly_reviews = df_cleaned.groupby(['review_year', 'review_month'])['number_of_reviews'].sum().reset_index()

pivot_reviews = monthly_reviews.pivot(index='review_month', columns='review_year', values='number_of_reviews')

plt.figure(figsize=(14, 8))

plt.subplot(2, 1, 1)
for year in pivot_reviews.columns:
    plt.plot(pivot_reviews.index, pivot_reviews[year], marker='o', label=year)

plt.title('Perbandingan Jumlah Review Bulanan Antar Tahun')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Review')
plt.xticks(range(1, 13), labels=month_abbr[1:13])
plt.legend(title='Tahun')
plt.grid(True)

plt.subplot(2, 1, 2)
pivot_reviews.plot(kind='bar', ax=plt.gca())
plt.title('Distribusi Review per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Review')
plt.xticks(range(12), labels=month_abbr[1:13], rotation=0)
plt.legend(title='Tahun')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_cleaned,
                x='longitude',
                y='latitude',
                hue='price',       # Warna berdasarkan harga
                size='price',      # Ukuran berdasarkan harga
                sizes=(20, 200),   # Range ukuran titik
                palette='rocket_r',
                alpha=0.6)

plt.title('Penyebaran Harga Listing Airbnb')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_reviews.T, cmap='YlGnBu', annot=True, fmt='.0f')
plt.title('Heatmap Jumlah Review per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Tahun')
plt.show()
```
